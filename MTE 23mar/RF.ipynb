```python
import pandas as pd
import numpy as np
df = pd.read_excel('Auction_HK_2016-2020 final .xlsx')
```

```python
df=df.drop(['artist','title','created'],axis=1)
```

```python
df = pd.get_dummies(df)
```

```python
df=df.dropna()
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
pd.set_option('display.max_columns', None)
```

```python
df=df.drop(df[(df['sales_price']<25000)].index)
```

```python
df=df.drop(df[(df['sales_price']>300000)].index)
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
y = df["sales_price"]
x=df.drop(['sales_price','sales_price_log'],axis=1)
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.50, random_state=42)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 2000 decision trees
rf = RandomForestRegressor(n_estimators = 3000, random_state = 42)
# Train the model on training data
rf.fit(x_train,y_train);
```

```python
# Use the forest's predict method on the test data
y_pred = rf.predict(x_test)
# Calculate the absolute errors
errors = abs(y_pred - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 0), '%.')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print('MAE: ', mean_absolute_error(y_test, y_pred))
print('MSE: ', mean_squared_error(y_test, y_pred)) 
```

```python
random_forest_out_of_bag = RandomForestRegressor(oob_score=True)
random_forest_out_of_bag.fit(x_train, y_train)
print(random_forest_out_of_bag.oob_score_) 
```
