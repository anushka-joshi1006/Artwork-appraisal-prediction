```python
import pandas as pd
import numpy as np
df = pd.read_excel('Auction_HK_2016-2020 final .xlsx')
```

```python
df=df.drop(['artist','title','created'],axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.subplots(figsize=(18,16))
sns.heatmap(df.corr(), annot= True)
```

```python
from sklearn.model_selection import train_test_split
y = df["sales_price"]
x = df.drop(["sales_price",'sales_price_log'], axis = 1)
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.70, random_state=1)  
from sklearn.linear_model import LinearRegression
```

```python
x
```

```python
df.describe()
```

```python
y
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test).round(2)
```

```python
lr.score(x_train, y_train).round(2)
```

```python
y_test_pred = lr.predict(x_test)
```

```python
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_test_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_test_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_test_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_test_pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_test_pred), 2))
```
