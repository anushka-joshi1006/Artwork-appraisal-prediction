```python
df=df[['sales_price','estimate_range','auction_location','surface','auction_year','auction_month','auction_weekday']]
```

```python
plt.scatter(df['auction_weekday'],df['sales_price'])

plt.xlabel('Auction Weekday')
plt.ylabel('sales price')
plt.show()
     
```

```python
res = sns.barplot(df['auction_weekday'],df['sales_price'])
plt.show()
```

```python
sns.heatmap(df.corr(), cbar = True, linewidths = 0.5)
```

```python
sns.pairplot(df)
```

```python
res = sns.countplot(y='auction_location', data = df, palette='rocket')
```

```python
res = sns.countplot(y='auction_weekday', data = df, palette='rocket')
```

```python
res = sns.countplot(y='auction_year', data = df, palette='rocket')
```
